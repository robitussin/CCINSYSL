# Jeepney Fare

In this example, we will build a Simple Linear Regression model to study the linear relationship between **fuel price** and **jeepney fare**.


<img src="jeepney.gif" />

## Step 1: Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2: Create the dataset

```python
x = [1,2,3,4,5]
y = [2,4,5,4,5]

dict = {'fuelprice': x, 'fare': y}

df = pd.DataFrame(dict)
df.head()
```

## Step 3: Plot the data points

```python
df.plot(x='fuelprice',y='fare',kind='scatter')
```

## Step 4: Create the model

```python
# Get the all features in the fuelprice column
X = df['fuelprice'].values

# Get the all features in the fare column
y = df['fare'].values
```

```python
# Building the model
m = 0
b = 0

L = 0.0001  # The learning Rate
epochs = 1000  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    y_pred = m * X + b  # The current predicted value of Y

    D_m = (-2/n) * sum(X * (y - y_pred))  # Derivative wrt m

    D_b = (-2/n) * sum(y - y_pred)  # Derivative wrt c

    m = m - L * D_m  # Update m
    
    b = b - L * D_b  # Update b
    
print("Slope: ", m)
print("y-intercept: ", b)
```

```python
y_pred = m*X + b

plt.scatter(X, y) 
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color='red')  # regression line
plt.show()
```
