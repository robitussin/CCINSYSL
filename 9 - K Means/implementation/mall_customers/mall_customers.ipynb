# K-Means

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
```

```python
# Importing the dataset
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

We only need the annual income and spending score features

```python
# Using the .iloc function to retrieve the AnnualIncome and SpendingScore from the dataframe
X = df.iloc[:, [3, 4]].values
```

```python
plt.scatter(df.AnnualIncome, df.SpendingScore)
plt.xlabel("Spending Score")
plt.ylabel("Annual Income")
```

```python
X
```

#### Set the number of clusters

```python
clusters = 4
```

#### Pass the number of clusters to the KMeans class

```python
kmeans = KMeans(n_clusters=clusters)
```

#### Train the KMeans model

```python
kmeans.fit(X)
```

#### Use `predict()` to predict each cluster

```python
y_kmeans = kmeans.predict(X)
```

#### Plot the clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200);
```
